```python
# 1. Pokreni detekciju
!python detect.py \
  --weights runs/train/roboflow_model/weights/best.pt \
  --img 640 \
  --conf 0.25 \
  --source /content/drive/MyDrive/R.jpg \
  --save-txt

# 2. Pronađi najnoviji 'exp' folder s rezultatima
import os
import glob

detect_dir = 'runs/detect'
exp_folders = sorted(glob.glob(f'{detect_dir}/exp*'), key=os.path.getmtime, reverse=True)
latest_exp = exp_folders[0]
label_files = glob.glob(f'{latest_exp}/labels/*.txt')

# 3. Učitaj klasne nazive iz ispravnog data.yaml
import yaml

with open("/content/yolov5/Fruits-by-YOLO-1/data.yaml", "r") as f:
    data_yaml = yaml.safe_load(f)

class_names = data_yaml['names']

# 4. Ispiši detekcije
detected_classes = []

for label_file in label_files:
    with open(label_file, "r") as f:
        for line in f:
            class_id = int(line.strip().split()[0])
            detected_classes.append(class_names[class_id])

if detected_classes:
    print("✅ Detektovani objekti:", set(detected_classes))
else:
    print("⚠️ Nijedan objekat nije detektovan.")
```
